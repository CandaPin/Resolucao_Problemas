# 2- Pré-processamanto

A quantidade máxima de **peças em caixas** precisa ser:
<ul>
    <li>120.000 peças para o “Tipo de buffer” CX e </li>
    <li>100.000 peças para o “Tipo de buffer” EN.</li>
 </ul>

```python
df_prod_completo.columns = df_prod_completo.columns.str.replace(' ', '_')
```

```python
df_prod_completo['Cor'] = df_prod_completo['Cor'].fillna('0')
```

```python
df_prod_completo['pai_cor'] = df_prod_completo['Item_pai'] + '_' + df_prod_completo['Cor']
df_prod_completo['pai_cor']
```

```python
df_prod_completo['chave_loja_buffer_onda'] = df_prod_completo['Loja'] + '_' + df_prod_completo['Tipo_de_buffer'] + '_' + df_prod_completo['Classe_de_onda']
```

```python
df_prod_completo['Volume_unit'] = df_prod_completo['Comprimento_unit'] * df_prod_completo['Largura_unit'] * df_prod_completo['Altura_unit']
```

```python
df_completo_FIFO = df_prod_completo.sort_values(by=['Data_da_ordem_de_produção', 'Ordem_de_produção', 'Loja','Classe_de_onda', 'Tipo_de_buffer','pai_cor'])
```

```python
with pd.option_context('display.max_columns', None):
    display(df_completo_FIFO.head())
```

Quantidade total de peças por tipo de buffer

```python
df_completo_FIFO.loc[df_completo_FIFO['Tipo_de_buffer'] == 'EN', 'Peças'].sum()
```

```python
df_completo_FIFO.loc[df_completo_FIFO['Tipo_de_buffer'] == 'CX', 'Peças'].sum()
```

Observa-se que a base completa tem mais itens do que o limite de produção diário. Por isso, iremos reduzir a quantidae de itens.

## 2.1- Limite buffer

Selecionar o limite do buffer tipo CX

```python
soma = 0
indices = []
tipo_buffer = 'CX'
max = 120000

for index, value in df_completo_FIFO[df_completo_FIFO['Tipo_de_buffer'] == tipo_buffer]['Peças'].items():
    soma += value
    if soma == max:
        indices.append(index)
        break
    elif soma > max:
        soma -= value
        break
    else:
        indices.append(index)

df_limite_buffer_CX = df_completo_FIFO.loc[indices]
df_limite_buffer_CX['Peças'].sum()
```

Selecionar o limite do buffer tipo EN

```python
soma = 0
indices = []
tipo_buffer = 'EN'
max = 100000

for index, value in df_completo_FIFO[df_completo_FIFO['Tipo_de_buffer'] == tipo_buffer]['Peças'].items():
    soma += value
    if soma == max:
        indices.append(index)
        break
    elif soma > max:
        soma -= value
        break
    else:
        indices.append(index)

df_limite_buffer_EN = df_completo_FIFO.loc[indices]
df_limite_buffer_EN['Peças'].sum()
```

```python
df_limite_buffer = pd.concat([df_limite_buffer_CX, df_limite_buffer_EN], ignore_index=True)
df_limite_buffer
```

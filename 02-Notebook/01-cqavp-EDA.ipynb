### 2.2.3 Limitação da capacidade de peças por turno

A quantidade máxima de **peças em caixas** precisa ser:
<ul>
    <li>120.000 peças para o “Tipo de buffer” CX e </li>
    <li>100.000 peças para o “Tipo de buffer” EN.</li>
 </l>

```python
df_completo_FIFO = df_prod_completo.sort_values(by=['Data da ordem de produção', 'Loja','Classe de onda', 'Tipo de buffer','pai-cor'])
```

```python
with pd.option_context('display.max_columns', None):
    display(df_completo_FIFO.head())
```

Quantidade total de peças por tipo de buffer

```python
df_completo_FIFO.loc[df_completo_FIFO['Tipo de buffer'] == 'EN', 'Peças'].sum()
```

```python
df_completo_FIFO.loc[df_completo_FIFO['Tipo de buffer'] == 'CX', 'Peças'].sum()
```

Selecionar o limite do buffer tipo CX

```python
soma = 0
indices = []
tipo_buffer = 'CX'
max = 120000

for index, value in df_prod_completo[df_prod_completo['Tipo de buffer'] == tipo_buffer]['Peças'].items():
    soma += value
    if soma == max:
        indices.append(index)
        break
    elif soma > max:
        soma -= value
        break
    else:
        indices.append(index)

df_limite_buffer_CX = df_prod_completo.loc[indices]
df_limite_buffer_CX['Peças'].sum()
```

Selecionar o limite do buffer tipo EN

```python
soma = 0
indices = []
tipo_buffer = 'EN'
max = 100000

for index, value in df_prod_completo[df_prod_completo['Tipo de buffer'] == tipo_buffer]['Peças'].items():
    soma += value
    if soma == max:
        indices.append(index)
        break
    elif soma > max:
        soma -= value
        break
    else:
        indices.append(index)

df_limite_buffer_EN = df_prod_completo.loc[indices]
df_limite_buffer_EN['Peças'].sum()
```

```python
df_limite_buffer = pd.concat([df_limite_buffer_CX, df_limite_buffer_EN], ignore_index=True)
df_limite_buffer
```
